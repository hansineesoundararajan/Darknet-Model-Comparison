Feature Selection

```python
from sklearn.feature_selection import VarianceThreshold
data_numeric = data.select_dtypes(include=['number'])
selector = VarianceThreshold(threshold=0.01)
df_filtered = selector.fit_transform(data_numeric)

df_filtered = pd.DataFrame(df_filtered, columns=data_numeric.columns[selector.get_support()])
df = df_filtered
df.head(5)
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split

# Select only numerical features, excluding the target variable
X = df.drop(columns=['Label'])  # Dropping target directly (No need for select_dtypes if all features are numeric)
y = df['Label']  # Target variable

# Split the dataset (Stratify ensures balanced class distribution if classification task)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=99, stratify=y
)

print("Training set size:", X_train.shape)
print("Test set size:", X_test.shape)
```

```python
from sklearn.model_selection import train_test_split, cross_val_score, KFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd

# Convert y to categorical (choose one method)
y = (y > y.median()).astype(int)  # Binary classification (Low=0, High=1)
# y = pd.qcut(y, q=3, labels=[0, 1, 2])  # Multi-class (Low=0, Medium=1, High=2)
# y = y.round().astype(int)  # Rounds continuous values to integers

# Standardizing features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

for seed in [1, 42, 99, 123, 777]:
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=seed)
    knn_classifier = KNeighborsClassifier(n_neighbors=5, metric='euclidean')
    knn_classifier.fit(X_train, y_train)
    accuracy = knn_classifier.score(X_test, y_test)
    print(f"Random State {seed} â†’ Test Accuracy: {accuracy:.4f}")


# K-Fold Cross Validation
kf = KFold(n_splits=5, shuffle=True, random_state=99)

# Cross-validation accuracy
cv_accuracy_scores = cross_val_score(KNeighborsClassifier(n_neighbors=5, metric='euclidean'),
                                     X_train, y_train, cv=kf, scoring='accuracy')

print(f"Cross-Validation Accuracy Scores: {cv_accuracy_scores}")
print(f"Average Cross-Validation Accuracy: {np.mean(cv_accuracy_scores):.4f}")

# Train KNN classifier
knn_classifier = KNeighborsClassifier(n_neighbors=5, metric='euclidean')
knn_classifier.fit(X_train, y_train)

# Predicting on test set
y_pred = knn_classifier.predict(X_test)

# Model Evaluation
accuracy_test = accuracy_score(y_test, y_pred)
print(f"Test Set Accuracy: {accuracy_test:.4f}")

print("Classification Report:\n")
print(classification_report(y_test, y_pred))
```

```python
print(f"Train Accuracy: {knn_classifier.score(X_train, y_train):.4f}")
print(f"Test Accuracy: {knn_classifier.score(X_test, y_test):.4f}")
```
